```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
df = pd.read_excel("Lab task 4 regression.xlsx")
df
```

```python
df.drop(['S.No'],axis=1,inplace=True)
df
```

```python
df = df.replace({'company': {'ABBB': 0,'Oxon': 1,'Bashh':2,'palm':3,'TSC':4}})
df = df.replace({'Gender': {'F': 0,'M': 1}})
df = df.replace({'Qualification': {'B.E': 0,'M.E': 1}})
df = df.replace({'Area of work': {'programming': 0,'predective analysis': 1,'cloud server management':2}})
df
```

```python
X = df.iloc[:, 1:6].values
Y1 = df.iloc[:, 6].values
Y2 = df.iloc[:, 0].values
X_train, X_test, Y1_train, Y1_test = train_test_split(X,Y1,random_state=0, test_size=0.30)
X_train, X_test, Y2_train, Y2_test = train_test_split(X,Y2,random_state=0, test_size=0.30)
```
