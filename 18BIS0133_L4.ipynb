```python

import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt
df = pd.read_excel("Lab task 4 regression.xlsx")
df
```

```python
df.drop(['S.No'],axis=1,inplace=True)
df
```

```python
# replaced all the strings to numeric value
df = df.replace({'company': {'ABBB': 0,'Oxon': 1,'Bashh':2,'palm':3,'TSC':4}})
df = df.replace({'Gender': {'F': 0,'M': 1}})
df = df.replace({'Qualification': {'B.E': 0,'M.E': 1}})
df = df.replace({'Area of work': {'programming': 0,'predective analysis': 1,'cloud server management':2}})
df
```

```python
#Training all the data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
linear_regression=LinearRegression() #model for company
linear_regression.fit(df[['Age','Gender','Qualification','no of experience in years','Area of work']],df.company)
```

```python

linear_regression.predict([[25,1,0,0,0]]) 
#Sanjay,25 Y, M, B.E ,programming with 0 years experience Company prediction
```
